```python
# analysis_notebook.ipynb

import sys
import os
import pandas as pd
import nltk

# Add the path to the src folder where sentiment_analysis_1.py is located
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../src")))

# Download required NLTK data
nltk.download("punkt")
nltk.download("stopwords")

# Import the necessary functions from sentiment_analysis_1.py
from sentiment_analysis_1 import (
    load_data,
    perform_sentiment_analysis,
    plot_sentiment_distribution,
    plot_top_words_per_topic,
    perform_topic_modeling,
    plot_word_cloud,
)

# Load data
df = load_data("../data/raw_analyst_ratings.csv")  # Adjust the file path as necessary

# Perform Sentiment Analysis
df_sentiment = perform_sentiment_analysis(df)
df_sentiment[["headline", "sentiment"]].head()  # Display sentiment for each headline
plot_sentiment_distribution(df_sentiment)  # Plot sentiment distribution

# Perform Topic Modeling (LDA)
lda_model, topics = perform_topic_modeling(df, num_topics=5)
plot_word_cloud(df, lda_model)  # Plot a word cloud for topics
plot_top_words_per_topic(lda_model, num_topics=5)  # Plot top words for each topic


```
