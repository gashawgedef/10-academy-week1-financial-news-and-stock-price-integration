```python
import sys
import os
import pandas as pd

# Add the src folder to the Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../src")))

# Import functions from time_series_analysis.py
from time_series_analysis import (
    load_data,
    publication_frequency_over_time,
    analyze_publishing_times,
    articles_per_publisher,
    extract_unique_publisher_domains,
)

# Define the file paths for input and output
input_file_path = "../data/raw_analyst_ratings.csv"
output_file_paths = {
    "publication_count": "publication_frequency.csv",
    "publishing_times": "publishing_times.csv",
    "publisher_stats": "publisher_stats.csv",
    "publisher_domains": "publisher_domains.csv",
}

# Load the data
try:
    df = load_data(input_file_path)
    print("Dataset loaded successfully.")
except Exception as e:
    print(f"Error loading the dataset: {e}")
    sys.exit(1)  # Exit if there's an error loading the data

# Display the dataset columns to confirm it's loaded correctly
print("\nDataset Columns After Loading:")
print(df.columns)

# Time Series Analysis: Publication Frequency Over Time
try:
    print("\nPublication Frequency Over Time:")
    publication_count = publication_frequency_over_time(df)
    print(publication_count.head())  # Show the first few rows of publication count
except Exception as e:
    print(f"Error analyzing publication frequency: {e}")

# Time Series Analysis: Analyzing Publishing Times
try:
    print("\nPublishing Times Analysis:")
    publishing_times = analyze_publishing_times(df)
    print(publishing_times.head())  # Show the publishing times for articles
except Exception as e:
    print(f"Error analyzing publishing times: {e}")

# Publisher Analysis: Articles per Publisher
try:
    print("\nArticles Per Publisher:")
    publisher_stats = articles_per_publisher(df)
    print(publisher_stats.head())  # Display the articles per publisher
except Exception as e:
    print(f"Error analyzing articles per publisher: {e}")

# Publisher Analysis: Unique Publisher Domains
try:
    print("\nUnique Publisher Domains:")
    publisher_domains = extract_unique_publisher_domains(df)
    print(publisher_domains.head())  # Display unique publisher domains
except Exception as e:
    print(f"Error extracting unique publisher domains: {e}")

# Optionally, save the results to CSV
try:
    publication_count.to_csv(output_file_paths["publication_count"])
    publishing_times.to_csv(output_file_paths["publishing_times"])
    publisher_stats.to_csv(output_file_paths["publisher_stats"])
    publisher_domains.to_csv(output_file_paths["publisher_domains"])
    print("\nResults saved to CSV files.")
except Exception as e:
    print(f"Error saving results to CSV: {e}")
```
